##**Fundamental Particles**
---
Particle physics is currently described by a theory called the Standard Model. The Standard Model describes all matter and most forces at the most fundamental level. The only force not accounted for is gravity, and combining our understanding of particle physics with our understanding of gravity is one of the largest open problems in physics. Below, you can see the different particles described by the Standard Model.

###**Bosons and Fermions**
---
These particles are grouped together based on their properties. In the Standard Model, **Bosons** are particles that carry a force or interaction. For example, the Standard Model says that when two charged particles attract or repel each other because they exchange **photons**, which are the carrier particles for the electromagnetic force. There are similar descriptions for the other forces described by the Standard Model. The strong force is carried by the **gluon**, while the weak force is mediated by two particles: the **W boson** and the **Z boson**.

| <h4> **Force/Interaction** </h4>	| <h4> **Carrier Particle** </h4> 	|
|-------------------	|------------------	|
| <h5> Electromagnetism </h5> 	| <h5> Photon     </h5>      	|
| <h5> Strong Force     </h5> 	| <h5> Gluon      </h5>      	|
| <h5> Weak Interaction </h5> 	| <h5> W/Z Bosons </h5>      	|

In contrast, the **Fermions** in the Standard Model make up all matter in the universe. Protons, electrons, and neutrinos are all examples of fermions. The fermions are further divided into two groups based on the observation that only some fermions experience the strong force. **Quarks** are fermions that experience the strong force, while **leptons** are fermions that do not experience the strong force. Just like the electromagnetic force, the strong force has an associated charge called **color charge**.

###**Note on Bosons and Fermions**
---
Bosons and fermions are actually defined based on the spin of the particle, which allows us to classify composite particles as bosons and fermions as well. **Spin** is a quantum property that causes particles to behave as if they were spinning, even if they are not physically rotating. Bosons are defined as particles that have integer spin (0,1,2, etc.), while fermions are defined as particles that have half-integer spin (1/2, 3/2, 5/2, etc.). Understanding spin is not necessary to complete these activities, but it is important to note that because of this definition, some particles that are not considered "force carriers" are bosons.

###**Antimatter**
---

The last major component of the standard model is the existence of anti-matter. Each particle shown in the standard model has a corresponding antiparticle. Some particles are their own antiparticles, such as the photon, the gluon, and the Z boson. Other particles, such as quarks, electrons, and muons have distinct antiparticles. Antiparticles are nearly identical to normal particles, but they have opposite charge and color charge. Furthermore, if a particle and its antiparticle collide, they **annihilate** and destroy each other. For this reason, the most stable forms of matter are made entirely out of matter or entirely out of antimatter.

Each particle has its own symbol. For example, neutrinos are given the symbol $\nu$ (nu) and the up quark is given the symbol $u$. Antiparticles are given the same symbol, but are written with a bar over it. For example, an antineutrino is written as $\overline{\nu}$ and the anti-up quark is written as $\overline{u}$.

##**Composite Particles**
---
But why is the strong force so important? It turns out that the strong force is so strong that the particles that experience it (quarks) are always stuck together. The particles formed by gluing quarks together are called **hadrons**. There are three types of hadrons:


1.   **Baryons** are hadrons composed of three quarks. Examples of baryons are protons, neutrons, and lambda baryons.
2.   **Antibaryons** are hadrons composed of three antiquarks. Antibaryons are the antiparticles of baryons. Antiprotons, antineutrons, and antilambdas.
3.   **Mesons** are hadrons composed of one quark and one anti-quark. Examples of mesons include pions, kaons, and D mesons. The antiparticle of a meson is another meson.

The process of classifying particles relies on only a few properties. The `particles` variable contains a dataframe. Run the cell below to display the first few rows.

Write a python script that iterates over the rows of the `particles` dataframe and prints out
```
[Particle] is a [Classification]
```
where `[Particle]` is replaced with the name of the particle, and `[Classification]` is replaced with the appropriate label from the following list:


*   Boson
*   Fermion

If the particle is a hadron, print out:
```
[Particle] is also a [Classification2]
```
where `[Classification2]` is replaced with the appropriate label from the following list:



*   Baryon
*   Antibaryon
*   Meson




#####**Solution:**

````python
for row in particles.iloc:
  if int(2*row['spin']) % 2 == 0:
    print(row['Particle'] + " is a boson.")
  else:
    print(row['Particle'] + " is a fermion.")
  if len(row['quarks']) > 0:
    if len(row['quarks']) == 2:
      print(row['Particle'] + " is also a meson.")
    elif 'Anti' in row['quarks'][0]:
      print(row['Particle'] + " is also an antibaryon.")
    else:
      print(row['Particle'] + " is also a baryon.")
````
