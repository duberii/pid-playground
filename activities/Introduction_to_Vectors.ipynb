## **Scalars and Vectors**
---

There are a variety of properties that we can measure in particle detectors, but only some of these properties are useful for studying the detected particles. These properties can be divided into two categories: scalar properties and vector properties.

A **scalar** is simply a number, and a **scalar property** is a property that can be interpreted as a number. For example, charge is a scalar property because we describe it using a number. Other examples of scalar properties are mass, time, speed, energy, and distance.

A **vector**, on the other hand, is a quantity that has both a magnitude and a direction. For example, velocity is a vector because it has both a magnitude (speed) and a direction. Position can also be considered a vector since it has a magnitude (distance from the origin) and a direction. Other examples of vectors are force, momentum, and acceleration. Vectors are usually represented by arrows, since the magnitude of the vector can be represented by the length of the arrow, and the direction is conveyed by the direction of the arrow.

Vector_components.svg.png

Because we live in 3 dimensions, the vectors we deal with will have three components: an x component, a y component, and a z component. We write these components together as follows:
$$
\begin{bmatrix}
x\\
y\\
z
\end{bmatrix}\begin{matrix}
\leftarrow \mbox{x component}\\
\leftarrow \mbox{y component}\\
\leftarrow \mbox{z component}
\end{matrix}
$$
When we write a vector in this form, we call it a **column vector**.

##**Adding Vectors**
---
The reason vector and scalar properties are so important is because in particle physics experiments, several vector and scalar quantities are conserved. For example, total energy (a scalar property) and total momentum (a vector quantity) are both conserved. "Total" refers to the fact that we take the sum over all of the particles in the detector. The momentum of an individual particle is conserved as well,  but it turns out that this isn't super helpful for our data analysis.

To calculate the total momentum, we will need to know how to add vectors together. In general, this is fairly difficult. However, when we know all of the components of the vectors, we simply add the components of the vectors together. In terms of column vectors,
$$
\begin{bmatrix}
x\\
y\\
z
\end{bmatrix}+\begin{bmatrix}
a\\
b\\
c
\end{bmatrix}=\begin{bmatrix}
x+a\\
y+b\\
z+c
\end{bmatrix}
$$

##**Calculating Magnitudes**
---

In terms of the components of a vector, the formula for the **magnitude** (length) of a vector, denoted $|v|$, looks something like the Pythagorean Theorem:
$$
|v|=\sqrt{x^2 + y^2 + z^2}
$$
Note that the magnitude of a vector is always a scalar.

##**Rescaling Vectors**
---
Later on, we will find particles for which we can only directly measure the magnitude of momentum. Using other methods, we will be able to find a vector that points in the direction of the momentum, but how do we use the magnitude of momentum and the direction to create the momentum vector?

The answer is that we need to **rescale** the vector. By rescaling the vector, we can create a vector that points in the same direction and has the same magnitude as the momentum vector. In some sense, we are recreating the momentum vector from a magnitude and a direction.

Rescaling a vector is sometimes called **scalar multiplication**, because to rescale a vector, we multiply it by a number. When working with a column vector, we can rescale a vector by multiplying all of its components by the rescaling factor:

$$
a\begin{bmatrix}
x\\
y\\
z
\end{bmatrix}=\begin{bmatrix}
a*x\\
a*y\\
a*z
\end{bmatrix}
$$

```python
a = 3
x = x_original * a
y = y_original * a
z = z_original * a
plot_vector(x,y,z)
```

$$

$$
In this case, $av$ is the rescaled vector, $a$ is the rescaling factor, and $v$ is the original vector.

#####**Solution:**

Our goal is to obtain a vector with the same direction as the provided vector, but a magnitude of 3. We will need to find the right value to multiply the vector by. To find this value, we need to plug in the correct values into the provided expression:
$$ |av| = |a||v|$$
In this case, $av$, our rescaled vector, should be our momentum vector. Since we know that the momentum vector should have a magnitude of 3, we have:
$$ 3 = |a| |v|$$
In this case, $v$ is our initial vector, and we can use its components to calculate its magnitude:

```python
x = 5
y = -1
z = 8
magnitude = (x**2 + y**2 + z**2)**0.5
print(magnitude)
```

We now know the value of $|v|$ (approximately 9.5). Thus we can find the value of $a$ using the following equation:
$$
|a| = \frac{3}{|v|}
$$
We want to preserve the direction of the vector, so based on question 6, we want $a$ to be positive. Thus we multiply each component of the vector by 3 divided by the magnitude.

```python
p = 3
x_rescaled = x * p / magnitude
y_rescaled = y * p / magnitude
z_rescaled = z * p / magnitude
print(x)
print(y)
print(z)
```
